```python
# wap to find max no. form there using function

def ma(a,b,c):
    if a>b and a>c:
        print("a is greatest : ",a)
    elif b>a and b>c:
        print("b is geratest : ",b)
    elif c>a and c>b:
        print("c is geratest : ",c)
a=int(input("enter  : "))
b=int(input("enter  : "))
c=int(input("enter  : "))
ma(a,b,c)
```

```python
# wap to min no. form the three numbers:

def mi(a,b,c):
    if a<b and a<c:
        print("a is greatest : ",a)
    elif b<a and b<c:
        print("b is geratest : ",b)
    elif c<a and c<b:
        print("c is geratest : ",c)
a=int(input("enter  : "))
b=int(input("enter  : "))
c=int(input("enter  : "))
mi(a,b,c)
```

```python
# # wap to string is reversed or not

def rev(s):
    if s==s[::-1]:
        print("the string is not reversed!!")
    else:
        print("the string is reversed!!")
        

s="aba"
rev(s)


```

```python
# 6. wap to print all prime number b/w 10-50


def prime():
    for i in range(10,51):
        for j in range(i):
            if i%1==0 and i%i==0 and i%(j+1)!=0:
                print("prime",i)
    
    
    
prime()    
```

```python
# wap to find leap year


def leap(year):
    if year%4==0 and year%100!=0 or year%400==0:
        print("this is a leap year : ",year)
    else:
        print("not leap year : ",year)
        
year=int(input("enter the year : "))
leap(year)
```
